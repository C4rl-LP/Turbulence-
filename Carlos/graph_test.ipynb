```python
import numpy as np
import matplotlib.pyplot as plt

def bump_function(s, c=1.0):
    """
    Calcula a funo bump B(s) = exp(-c / (1 - s^2)) para |s| < 1
    e retorna 0 para |s| >= 1.
    """
    # Cria um array de zeros com o mesmo tamanho de s
    B = np.zeros_like(s)
    
    # Encontra os ndices onde a condio |s| < 1  verdadeira
    valid_indices = np.abs(s) < 1.0
    
    # Aplica a frmula apenas nos ndices vlidos para evitar diviso por zero
    B[valid_indices] = np.exp(-c / (1.0 - s[valid_indices]**2))
    
    return B

# Define o domnio de s (de -1.5 a 1.5 para mostrar o decaimento a zero nas bordas)
s_values = np.linspace(-1.5, 1.5, 1000)

# Valores de c para comparar
c_values = [0.7, 1.0, 2.0, 5.0]

# Configurao do grfico
plt.figure(figsize=(10, 6))

for c in c_values:
    B_values = bump_function(s_values, c)
    plt.plot(s_values, B_values, label=f'c = {c}', linewidth=2)

# Estilizao do grfico
plt.title('Funo Bump $B(s)$ para diferentes valores de c', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('B(x)', fontsize=12)
plt.axhline(0, color='black', linewidth=1) # Eixo x
plt.axvline(0, color='black', linewidth=1) # Eixo y
plt.axvline(-1, color='red', linestyle='--', alpha=0.5, label='Fronteiras |x| = 1')
plt.axvline(1, color='red', linestyle='--', alpha=0.5)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)

# Exibe o grfico
plt.show()
```
